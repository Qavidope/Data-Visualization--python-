#  Dealing with missing value 

```python
dfnew2=pd.read_csv('baseball.csv',usecols=['Tm','Playoffs','Attendance'])
dfnew2.head(13)
```

```python
dfnew2['Playoffs'].fillna(0,inplace=True)
dfnew2.head(10)
dfnew2.isnull().sum()
```

```python
dfnew2.head(13)
```

```python
dfnew3=pd.read_csv('baseball.csv',usecols=['Tm','Playoffs','Attendance'])
dfnew3.head(9)
```

```python
dfnew3.fillna(method='ffill',inplace=True)
dfnew3.head(10)
```

```python
dfnew3.fillna(method='bfill',inplace=True)
dfnew3.head(10)
```
