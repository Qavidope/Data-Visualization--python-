```python
import numpy as np
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
```

```python
titanic=pd.read_csv('train.csv')
titanic.head()
```

```python
titanic.drop(labels=['PassengerId','Name','Ticket','Cabin',"Embarked"],axis=1,inplace=True)
titanic.head()
```

```python
titanic.dtypes 
```

```python
convertnumeric={'male':1,"female":0}
titanic.Sex=titanic.Sex.map(convertnumeric)
titanic.head()
```

```python
titanic.dtypes
```

```python
titanic.isnull().sum()
```

```python
titanic.shape
```

```python
titanic.fillna(value=titanic.Age.mean(),inplace=True)
titanic.isnull().sum()
```

```python
titanic.head(10)
```

```python
tree =DecisionTreeClassifier()
survived=titanic['Survived'] #we need to take out Survived column as survived 
survived.head()
```

```python
titanic.drop(['Survived'],axis=1,inplace=True)
titanic.head()
```

```python
titanic.shape
```

```python
tree.fit(X=titanic[:800],y=survived[:800])        # fit means training   slicing from 0 to 800
tree.score(X=titanic[800:],y=survived[800:])      # score means 
```

```python
me=np.array([9,1,99,1,0,8])
me=me.reshape(1,-1)
tree.predict(me)
```
